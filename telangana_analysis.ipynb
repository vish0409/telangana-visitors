```python
import pandas as pd
import os
import numpy as np
```

```python
# Concatinating the domestic visitors datasets from 2016 to 2019
files=[file for file in os.listdir('./domestic_visitors')]
domestic_data=pd.DataFrame()
for file in files:
    df=pd.read_csv('./domestic_visitors/'+file)
    domestic_data=pd.concat([domestic_data,df])
display(domestic_data)
```

```python
# Concatinating the foreign visitors datasets from 2016 to 2019
files=[file for file in os.listdir('./foreign_visitors')]
foreign_data=pd.DataFrame()
for file in files:
    df=pd.read_csv('./foreign_visitors/'+file)
    foreign_data=pd.concat([foreign_data,df])
display(foreign_data)
```

```python
#domestic_data.drop(['date'],axis=1,inplace=True)
#foreign_data.drop(['date'],axis=1,inplace=True)
```

```python
# Checking all blank and null rows in domestic data
nan_df=domestic_data[domestic_data['visitors'].isna()]
display(nan_df.head())
blank_df=domestic_data[domestic_data['visitors']==" "]
display(blank_df)
```

```python
domestic_data=domestic_data.replace(" ",np.nan)
domestic_data=domestic_data.replace("0",np.nan)
domestic_data.dropna(inplace=True)
```

```python
# Checking all blank and null rows in foreign data
nan_df=foreign_data[foreign_data['visitors'].isna()]
display(nan_df.head())
blank_df=foreign_data[foreign_data['visitors']==" "]
display(blank_df)
zero_df=foreign_data[foreign_data['visitors']=="0"]
display(zero_df)
```

```python
foreign_data=foreign_data.replace(" ",np.nan)
foreign_data=foreign_data.replace("0",np.nan)
foreign_data.dropna(inplace=True)
```

```python
display(domestic_data.dtypes)
display(foreign_data.dtypes)
```

```python
# Setting the datatypes to their correct format

domestic_data['visitors']=pd.to_numeric(domestic_data['visitors'])
display(domestic_data.head())
foreign_data['visitors']=pd.to_numeric(foreign_data['visitors'])
display(foreign_data.head())
```

```python
foreign_data.to_excel("foreign_data.xlsx",index=False)
domestic_data.to_excel("domestic_data.xlsx",index=False)
```
